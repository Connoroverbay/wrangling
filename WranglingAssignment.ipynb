```python
import pandas as pd
import numpy as np
import os
os.getcwd()
```

```python
df = pd.read_csv('/content/airbnb_hw.csv', low_memory=False)
df.head(10)
```

```python
df['Price'].unique()
prices = df['Price']
prices = prices.str.replace(',','')
prices = pd.to_numeric(prices)
prices.unique()
print(np.mean(prices))
```

The changes made in this section were to remove all of the commas inside of the prices category for numbers over 1000 so that the variable could by typecasted into a numeric variable type and used for computational purposes in data analysis. The mean command at the end was to check and make sure that the typecast was successful. The code would return an error if the variable was still a string. This is also more effective than other methods because it does not require removing any data values.

```python
df = pd.read_csv('/content/sharks.csv', low_memory=False)
df.head(10)
```

```python
df['Type'].value_counts()
types = df['Type']
types = types.replace(['Questionable','Unconfirmed','Unverified','Under investigation','Invalid'],'Unknown')
types = types.replace(['Sea Disaster','Boating','Boat','Watercraft','Boatomg'],'Water Incident')
df['Type']=types
df['Type'].value_counts()
```

The data seemed to have too many similar categorical options for the Type of incident so I combined some of the similar ones into larger categories that will be more helpful for analysis. Also one of the subcategories was a clear typo so I included it in the larger category if it was spelled right.

```python
url = 'http://www.vcsc.virginia.gov/pretrialdataproject/October%202017%20Cohort_Virginia%20Pretrial%20Data%20Project_Deidentified%20FINAL%20Update_10272021.csv'
df = pd.read_csv(url,low_memory=False)
df.head(10)
```

```python
df.columns.to_list()
```

```python
release = df['WhetherDefendantWasReleasedPretrial']
release = release.replace(0,"No")
release = release.replace(1,"Yes")
release = release.replace('', np.nan)
df['WhetherDefendantWasReleasedPretrial'].rename("Released Pretrial")
df['Released Pretrial'] = release
df.head()
df['Released Pretrial'].value_counts()
```
